```python
# Load the input image
img = cv2.imread("/content/drive/MyDrive/My PC dell/Study and Research/Dr bhuiyan/TA/spring 2023/Cartonize/roi.jpg")
```

```python
# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
# Apply median blur to the grayscale image
gray = cv2.medianBlur(gray, 5)
cv2_imshow(gray)
```

```python
# Detect edges in the image using adaptive thresholding
edges = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9)
cv2_imshow(edges)
```

```python
# Reduce the color palette of the input image
color = cv2.bilateralFilter(img, 9, 300, 300)

# Combine the edges and color images to create the final cartoon image
cartoon = cv2.bitwise_and(color, color, mask=edges)

# Save the output image
cv2.imwrite("cartoon_image.jpg", cartoon)
```
